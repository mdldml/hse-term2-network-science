Again, the idea of the model:

1. Start with regular lattice with N nodes, K nearest neighbours
2. For each node take every edge and rewire it with probability $p$, assuming that there is no loops and edge duplications

```python
def gen_regular_graph(n, k):
    G = nx.Graph()
    nodes = list(range(n))
    for j in range(1, k // 2 + 1):
        targets = nodes[j:] + nodes[0:j] # first j nodes are now last in list
        G.add_edges_from(zip(nodes,targets))
    return G

def degree_distr(k, bins = 100):
    yh, binEdges = np.histogram(k, bins)
    bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
    return yh, bincenters
```

1.
Implementation of rewind function:

```python
def rewind(G, p):
    result = G.copy()
    
    for node in result.nodes():
        for edge in result.edges([node]):
            if np.random.random_sample() < p:
                u, v0 = edge
                v1 = np.random.choice(result.nodes())
                result.remove_edge(u, v0)
                result.add_edge(u, v1)
                
    return result
```

Example of regular graph:

```python
nx.draw_circular(gen_regular_graph(30, 10))
```

And of rewired one:

```python
nx.draw_circular(rewind(gen_regular_graph(30, 10), 0.1))
```

Average path length function:

```python
def path_length_avg(G):
    sample_size = 100
    chosen_shortest_paths = np.array([])
    
    for v in np.random.choice(G, sample_size):
        paths = nx.single_source_shortest_path_length(G, v)
        chosen_shortest_paths = np.concatenate((chosen_shortest_paths, np.random.choice(list(paths.values()), sample_size)))
    
    return np.mean(chosen_shortest_paths)
```

Dependency of average path length on p is shown below (we average results over 5 iterations for each p).  

```python
G = gen_regular_graph(n, 6)
results = [np.mean([path_length_avg(rewind(G, p)) for _ in range(5)]) for p in ps]
plt.loglog(ps, results)
```

Similar dependency of transitivity on $p$:

```python
G = gen_regular_graph(n, 6)
results = [np.mean([nx.transitivity(rewind(G, p)) for _ in range(5)]) for p in ps]
plt.loglog(ps, results)
```

One can notice that transitivity drops later than short path length.

We use the following functions to draw the degree distributions:

```python
def get_freqs(degrees):
    freqs = {}
    
    for degree in degrees:
        freqs.setdefault(degree, 0)
        freqs[degree] += 1.0 / len(degrees)
        
    return freqs

def get_binned_freqs(degrees, base=2):
    freqs = get_freqs(degrees)
    freqs_binned = {}
    
    for degree in freqs.keys():
        f_bin_upper = 1 if degree < 1 else base ** (ceil(log(degree, base))) 
        f_bin = (0.5 * f_bin_upper) if (f_bin_upper == 1) else ((1 + base) * f_bin_upper)
        freqs_binned.setdefault(f_bin, 0)
        freqs_binned[f_bin] += freqs[degree]
        
    return freqs_binned
```

Several degree distributions on one plot for different value of $p$:

```python
G = gen_regular_graph(n, 6)
for p in ps:
    H = rewind(G, p) #rewind(G, p)
    degrees = sorted([H.degree(v) for v in H.nodes()])
    freqs_binned = get_binned_freqs(degrees, base=1.1)
    plt.loglog(sorted(freqs_binned.keys()), [freqs_binned[key] for key in sorted(freqs_binned.keys())])  
```

One can note that the peak ($Pr(k_i = k)$) dissolves when increasing $p$.

```python
H = G.subgraph([n for n in G.nodes() if G.out_degree(n) > 10])
layout = nx.spring_layout(H, k=3 / sqrt(len(H.nodes())))
full_layout = nx.spring_layout(G, k=3 / sqrt(len(H.nodes())))
```

The following function takes graph, its subgraph for plotting, calculated centrality dictionary, layout, and prints out top nodes for that centrality metrics together with the graph plot.

```python
def plot_graph(G, H, dc, layout, print_top=True, color_nodes=True, base_node_size=500):
    plt.figure(1, figsize=(16, 9))
    
    H_nodes = set(H.nodes())
    top_central_nodes_list = sorted(dc.items(), key=lambda x: x[1], reverse=True)[:int(0.03 * len(dc))]
    top_central_nodes = set(map(lambda x: x[0], top_central_nodes_list))
    
    labels_map = {}
    for i, node in enumerate(top_central_nodes_list):
        labels_map[node[0]] = i

    if print_top:
        for node, score in top_central_nodes_list:
            print("%s - %s (score: %f), %s %s" % (labels_map[node], G.node[node]['label'], score, ('con' if G.node[node]['value'] else 'lib'), ('' if node in H_nodes else '(NOT SHOWN)')))
       
    node_colors = ['red' if H.node[n]['value'] else 'blue' for n in H.nodes()] if color_nodes else 'red'
    
    nx.draw_networkx(H,
                     layout,
                     nodelist=H.nodes(),
                     node_size=[(4 * base_node_size * dc[n] if n in top_central_nodes else base_node_size * dc[n]) for n in H.nodes()],
                     node_color=node_colors,
                     edgelist=[edge for edge in H.edges()],
                     #edge_color=[('magenta' if ((edge[0] in top_central_nodes) and (edge[1] in top_central_nodes)) else 'black') for edge in H.edges()], 
                     width=0.05,
                     labels = {n : (labels_map[n] if n in labels_map else '') for n in H.nodes()},
                     font_size=8)
```

Let's begin with the out degree centrality (most cited (directly) blogs): 

```python
plot_graph(G, H, nx.out_degree_centrality(G), layout)
```

Let's also see how the vertices with the highest degree interact with each other:

```python
dc = nx.out_degree_centrality(G)
top_central_nodes_list = sorted(dc.items(), key=lambda x: x[1], reverse=True)[:int(0.03 * len(dc))]
top_central_nodes = set(map(lambda x: x[0], top_central_nodes_list))

H1 =G.subgraph(top_central_nodes)
plot_graph(G, H1, dc, nx.spring_layout(H1))
```

We see that there appear two almost separate clusters, liberal blogs and conservative blogs, and they rarely cite each other.

We continue with eigenvector centrality (full graph is shown since many of most central nodes do not present in the subgraph).

We see that this metric mostly selects neighbours as the top central nodes. And, of course, since liberals and conservatives rarely cite each other, almost all nodes are of same direction (liberal). The only conservative blog, (24), instapundit, is often considered a libertarian blog and coincides with liberals on some issues (see https://en.wikipedia.org/wiki/Instapundit).

Let's continue with katz centrality.

```python
plot_graph(G, G, nx.katz_centrality(G, max_iter=1000, alpha=0.02), nx.spring_layout(G))
```

As we see, this metric gives us more meaningfull set of top blogs.

Function to generate Erdos model graph:

```python
def randgp(n, p):
    G = nx.Graph()
    G.add_nodes_from(range(0, n))
    
    for i in range(0, n - 1):
        for j in range(i + 1, n):
            if rnd.random() <= p:
                G.add_edge(i,j)
        
    return G
```

Let's start with Small world model and check average length path and transitivity for different values of p.

```python
ps = 10 ** np.arange(-5, 0, 0.3)
GG = gen_regular_graph(len(G.nodes()), 13)
results = [np.mean([path_length_avg(rewind(GG, p)) for _ in range(5)]) for p in ps]
print(results)
```
